```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter


```

```python
df = pd.read_csv('datos-tipo-cambio-usd-futuro-dolar-frecuencia-diaria.csv', index_col='indice_tiempo')
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
nestork = df.loc['2003-05-25':'2007-12-10']
nestork['usd_promedio'] = nestork['tipo_cambio_a3500'].mean()
```

```python
cristina1 = df.loc['2007-12-10':'2011-12-9']
cristina1['usd_promedio'] = cristina1['tipo_cambio_a3500'].mean()
```

```python
cristina2 = df.loc['2011-12-10':'2015-12-09']
cristina2['usd_promedio'] = cristina2['tipo_cambio_a3500'].mean()
```

```python
macri = df.loc['2015-12-10':'2019-12-09']
macri['usd_promedio'] = macri['tipo_cambio_a3500'].mean()
```

```python
sns.set()
sns.set_style("whitegrid")

plt.figure(figsize=(11.69, 8.27))
plt.plot(nestork.index, nestork['tipo_cambio_a3500'], label='Financiera', color='#F39C12')
plt.plot(nestork.index, nestork['usd_promedio'], label='Promedio financiera', linestyle='--', alpha=0.35, linewidth=2.5, color='#F39C12')

#plt.title('Total deudas financiera y comercial por ao (2007-2017)', fontdict={'fontsize': 20}, loc='center')
#plt.xlabel('Ao', size=16)
#plt.ylabel('Monto (en millones de U$S)', size=16)

plt.legend()

#plt.tight_layout()
plt.xlim('2003-05-25', '2007-12-10')


plt.show()
```
